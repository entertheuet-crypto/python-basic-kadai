```python
class Human:
    def __init__ (self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            return f"{self.name}は大人です。"

        else:
            return f"{self.name}は大人ではありません。"

person1 = Human("一郎", 30)
person2 = Human("二郎", 25)
person3 = Human("三郎", 15)

people = [person1, person2, person3]

for person in people:
    print(person.check_adult())
```
